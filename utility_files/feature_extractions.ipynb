```python
import numpy as np
from scipy.io.wavfile import read
import matplotlib.pyplot as plt
import os
import pandas as pd
import librosa
```

```python
def extract_audio_features(audio_path):
    y, sr = librosa.load(audio_path, sr = 22050)
    tempo, _ = librosa.beat.beat_track(y=y, sr=sr) #estimates global tempo

    #harmonic to noise ratio (HNR)
    harmonic, percussive = librosa.effects.hpss(y)
    hnr = np.mean(librosa.power_to_db(np.abs(harmonic)) - librosa.power_to_db(np.abs(percussive)))

    #spectral centroid
    spectral_centroid = librosa.feature.spectral_centroid(y=y, sr=sr)

    #spectral bandwidth
    spectral_bandwidth = librosa.feature.spectral_bandwidth(y=y, sr=sr)

    #spectral flatness
    spectral_flatness = librosa.feature.spectral_flatness(y=y)

    return [int(tempo), np.mean(hnr), np.mean(spectral_centroid), 
            np.mean(spectral_bandwidth), np.mean(spectral_flatness)]
```

```python
column_names = ['Tempo', 'HNR', 'Spectral_Centroid', 'Spectral_Bandwith', 'Spectral_Flatness']
variables = pd.DataFrame(columns = column_names)
```

```python
for filename in os.listdir('/Users/rgu/Desktop/UROPs/UROP4/diego_100_audios'):
    if filename.lower().endswith('.mp3'):
        new_row = extract_audio_features(f'/Users/rgu/Desktop/UROPs/UROP4/diego_100_audios/{filename}')
        variables.loc[len(variables)] = new_row
```
