```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
raw_diego_46 = pd.read_csv("/Users/rgu/Desktop/UROPs/UROP4/raw_diego_46.csv")
#dry_weight is straight from the bag
#weight0 is after submerging in water for 1 minute
#weight1 is after 1 minute of the song
#weight2 is after another minute of the same song looped
raw_diego_46['song_id'] = raw_diego_46['song_id'].astype('category')
raw_diego_46['water_loss1'] = raw_diego_46['weight0'] - raw_diego_46['weight1']
raw_diego_46['water_loss2'] = raw_diego_46['weight1'] - raw_diego_46['weight2']
display(raw_diego_46)
```

```python
#sanity check
print((raw_diego_46['weight0'][0] + raw_diego_46['weight0'][1])/2)
print((raw_diego_46['weight0'][2] + raw_diego_46['weight0'][3])/2)

print((raw_diego_46['weight1'][0] + raw_diego_46['weight1'][1])/2)
print((raw_diego_46['weight1'][2] + raw_diego_46['weight1'][3])/2)
```

```python
diego_46 = pd.DataFrame()

diego_46['song_id'] = raw_diego_46['song_id'].iloc[::2].values
diego_46['avg_weight0'] = (raw_diego_46['weight0'][::2].values + raw_diego_46['weight0'][1::2].values)/2
diego_46['avg_weight1'] = (raw_diego_46['weight1'][::2].values + raw_diego_46['weight1'][1::2].values)/2
diego_46['avg_water_loss'] = diego_46['avg_weight0'] - diego_46['avg_weight1']


diego_46['song_id'] = diego_46['song_id'].astype('category')
display(diego_46)
```

```python
diego_46.to_csv("/Users/rgu/Desktop/UROPs/UROP4/diego_46.csv", index=False)
```

```python
#One-Way ANOVA for first minute
model1 = ols('water_loss1 ~ C(song_id)', data=raw_diego_46).fit()

anova_results1 = sm.stats.anova_lm(model1, typ=2)
print(anova_results1)
```

```python
#One-Way ANOVA for second minute
model2 = ols('water_loss2 ~ C(song_id)', data=raw_diego_46).fit()

anova_results2 = sm.stats.anova_lm(model2, typ=2)
print(anova_results2)
```

```python
audio_variables = pd.read_csv("/Users/rgu/Desktop/UROPs/UROP4/audio_variables.csv")
display(audio_variables)
```

```python
repeatedfreq = [val for val in audio_variables['fundamental_frequency'] for _ in range(2)]
repeatedamp = [val for val in audio_variables['RMS_amplitude'] for _ in range(2)]


raw_diego_46['fundamental_frequency'] = repeatedfreq[0:94]
raw_diego_46['RMS_amplitude'] = repeatedamp[0:94]
display(raw_diego_46)
```

```python
#two-way ANOVA for first minute
model = ols('water_loss1 ~ C(fundamental_frequency) + C(RMS_amplitude)', data=raw_diego_46).fit()

anova_results = sm.stats.anova_lm(model, typ=2)
print(anova_results)

```

```python
#two-way ANOVA for second minute
model = ols('water_loss2 ~ C(fundamental_frequency) + C(RMS_amplitude)', data=raw_diego_46).fit()

anova_results = sm.stats.anova_lm(model, typ=2)
print(anova_results)
```
